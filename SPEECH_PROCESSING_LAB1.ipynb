```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
y, sr = librosa.load("C:/Audio files/SPEECH_PROCESSING_AUDIO.wav")
librosa.display.waveshow(y)
```

```python
duration_seconds = len(y) / sr  
amplitude_range = max(abs(y)) - min(abs(y))

print(f"Signal Length: {duration_seconds} seconds")
print(f"Magnitude Range: {amplitude_range}")
```

```python
from IPython.display import Audio

segment_start = 0.8
segment_end = 1.8

audio_segment = y[int(segment_start * sr):int(segment_end * sr)]

Audio(y, rate=sr)
```

```python
from IPython.display import Audio, display


segment_1_start_time = 0.5
segment_1_end_time = 1.5

segment_2_start_time = 1.6
segment_2_end_time = 2.6

abrupt_segment_start_time = 1.0
abrupt_segment_end_time = 2.0

segment_1_audio = y[int(segment_1_start_time * sr):int(segment_1_end_time * sr)]
segment_2_audio = y[int(segment_2_start_time * sr):int(segment_2_end_time * sr)]
abrupt_segment_audio = y[int(abrupt_segment_start_time * sr):int(abrupt_segment_end_time * sr)]

print("Playing Segment 1:")
display(Audio(segment_1_audio, rate=sr))

print("Playing Segment 2:")
display(Audio(segment_2_audio, rate=sr))

print("Playing Abruptly Segmented Speech:")
display(Audio(abrupt_segment_audio, rate=sr))
```
