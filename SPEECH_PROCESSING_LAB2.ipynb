```python
#A1

import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd

audio_signal, sample_rate = librosa.load("C:/Users/VIGNESH/Downloads/SPEECH_PROCESSING_AUDIO.wav")

first_derivative = np.diff(audio_signal)
first_derivative /= np.max(np.abs(first_derivative))  # Normalize

print("Playing First Derivative Signal:")

plt.figure(figsize=(10, 5))
librosa.display.waveshow(first_derivative, sr=sample_rate, color='black')
plt.title('FIRST DERIVATIVE')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(first_derivative, rate=sample_rate)
```

```python
plt.figure(figsize=(10, 5))
librosa.display.waveshow(audio_signal, sr=sample_rate, color='blue')
plt.title('Waveform')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
plt.show()

ipd.Audio(audio_signal, rate=sample_rate)
```

```python
#A2
zero_crossings = np.where(np.diff(np.sign(first_derivative)))[0]
zero_crossing_diffs = np.diff(zero_crossings)
threshold = 1000

# Distinguish between speech and silence based on the threshold
speech_intervals = zero_crossing_diffs[zero_crossing_diffs > threshold]
silence_intervals = zero_crossing_diffs[zero_crossing_diffs <= threshold]

# Calculate average lengths
avg_speech_interval_length = np.mean(speech_intervals)
avg_silence_interval_length = np.mean(silence_intervals)

# Print average lengths
print("Average length between consecutive zero crossings in speech regions:", avg_speech_interval_length)
print("Average length between consecutive zero crossings in silence regions:", avg_silence_interval_length)

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(zero_crossing_diffs, label='All intervals', color='red')
plt.plot(np.arange(len(speech_intervals)), speech_intervals, 'ro', label='Speech intervals', color='red')
plt.plot(np.arange(len(speech_intervals), len(speech_intervals) + len(silence_intervals)), silence_intervals, 'bo', label='Silence intervals', color='black')
plt.title('Pattern of Zero Crossings')
plt.xlabel('Interval Index')
plt.ylabel('Difference between Consecutive Zero Crossings')
plt.legend()
plt.show()

# Print patterns of zero crossings
print("Pattern of Zero Crossings:")
print("All intervals:", zero_crossing_diffs)
print("Speech intervals:", speech_intervals)
print("Silence intervals:", silence_intervals)
```

```python
#A3 
import librosa
import numpy as np
import matplotlib.pyplot as plt

my_word_files = ['Record-001-me.wav', 'Record-002-me.wav', 'Record-003-me.wav', 'Record-004-me.wav', 'Record-005-me.wav']
teammate_word_files = ['Record-001.wav', 'Record-002.wav', 'Record-003.wav', 'Record-004.wav', 'Record-005.wav']
word_labels = ['mathematics_for_intelligent_systems', 'natural_language_processing', 'career_growth_oppurtunities', 'I_am_in_delhi', 'mudumba_krishna_hanumant_vignesh']
my_word_durations = []
teammate_word_durations = []

for word_file in my_word_files:
    audio_data, sample_rate = librosa.load(word_file, sr=None)
    duration_secs = len(audio_data) / sample_rate
    my_word_durations.append(duration_secs)

for word_file in teammate_word_files:
    audio_data, sample_rate = librosa.load(word_file, sr=None)
    duration_secs = len(audio_data) / sample_rate
    teammate_word_durations.append(duration_secs)

print("Lengths of the spoken words MINE:", my_word_durations)
print("Lengths of the spoken words TeamMate:", teammate_word_durations)

plot_bar_width = 0.35
plot_indices = np.arange(len(word_labels))

plt.figure(figsize=(12, 6))
plt.bar(plot_indices - plot_bar_width/2, my_word_durations, plot_bar_width, label='My Words', color='skyblue')
plt.bar(plot_indices + plot_bar_width/2, teammate_word_durations, plot_bar_width, label="Teammate's Words", color='orange')
plt.xlabel('Words')
plt.ylabel('Length (seconds)')
plt.title('Comparison of Spoken Word Durations')
plt.xticks(plot_indices, word_labels)
plt.legend()

plt.show()
```
